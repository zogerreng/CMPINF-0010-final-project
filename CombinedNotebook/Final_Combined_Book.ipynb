# Walkability

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

pittmap = gpd.read_file("walkscorescombined/WalkScores.shp")
#hoods = pd.read_csv("walkscores.csv")
pittmap = pittmap.sort_values(by=['Walk_Score'], ascending=False)

print(pittmap[['NAME','Walk_Score']].head(16))
```

```python
var = 'Walk_Score'

vmin, vmax = 0, 100

fig, ax = plt.subplots(1, figsize=(20, 12))
ax.axis('off')
ax.set_title('Walkability Scores in Allegheny County', fontdict={'fontsize': '25', 'fontweight' : '3'})

ax.margins(x=-0.30,y=-0.30)

sm = plt.cm.ScalarMappable(cmap='Greens', norm=plt.Normalize(vmin=vmin, vmax=vmax))

cbar = fig.colorbar(sm)

pittmap.plot(column=var, cmap='Greens', linewidth=0.8, ax=ax, edgecolor='0.8')

pittmap.apply(lambda x: ax.annotate(text=x.NAME, xy=x.geometry.centroid.coords[0], ha='center'),axis=1);
```

## Walkability Data Set Conclusion 
Each polygon on the map corresponds to a separate census tract. The cluster of census tracts 706, 705, 703, and 708 correspond to the Shadyside neighborhood. In the top 15 entries, there were other tracts with high walkscores, but they were all more isolated and weren't representative of entire neighborhoods. Squirrel Hill, the neighborhood we originally were targeting, had one tract appear in the top 15. Census tract 1408 was that section, and it had a respectable walkability score of 88. However, the rest of the neighborhood was not as consistently high as Shadyside. Thus, because no other neighborhood was as walkable, I found it easy to conclude that Shadyside was the best neighborhood to live in (based on walkability, at least.) Being able to walk to many amenities in a neighborhood yields many benefits, such as improved health and improved air quality.

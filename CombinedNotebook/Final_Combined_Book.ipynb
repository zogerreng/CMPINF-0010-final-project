In conclusion, I have to give best neighborhood to Greenfield as I would spend almost no money in meter fees and I would have a very low chance of receiving a parking ticket if I chose not to pay.

Walking 

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

pittmap = gpd.read_file("walkscores/WalkScores.shp")
#hoods = pd.read_csv("walkscores.csv")
pittmap = pittmap.sort_values(by=['Walk_Score'], ascending=False)

print(pittmap[['NAME','Walk_Score']].head(16))
```

```python
var = 'Walk_Score'

vmin, vmax = 0, 100

fig, ax = plt.subplots(1, figsize=(20, 12))
ax.axis('off')
ax.set_title('Walkability Scores in Allegheny County', fontdict={'fontsize': '25', 'fontweight' : '3'})

ax.margins(x=-0.30,y=-0.30)

sm = plt.cm.ScalarMappable(cmap='Greens', norm=plt.Normalize(vmin=vmin, vmax=vmax))

cbar = fig.colorbar(sm)

pittmap.plot(column=var, cmap='Greens', linewidth=0.8, ax=ax, edgecolor='0.8')

pittmap.apply(lambda x: ax.annotate(text=x.NAME, xy=x.geometry.centroid.coords[0], ha='center'),axis=1);
```

Conclusion 
The best neighborhood based on wlaking score is Shadyside. Shadyside offers a close community in many ways. You are easily connected with other people, restaurants, parks, and other amentities. The central location of Shadyside definitely contributes to its overall best score.
